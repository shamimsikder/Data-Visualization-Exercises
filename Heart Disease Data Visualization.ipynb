# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## There are 14 columns in the dataset, which are described below.

1. Age: displays the age of the individual.

2. Sex: displays the gender of the individual using the following format :
1 = male
0 = female

3. Chest-pain type: displays the type of chest-pain experienced by the individual using the following format :
1 = typical angina
2 = atypical angina
3 = non — anginal pain
4 = asymptotic

4. Resting Blood Pressure: displays the resting blood pressure value of an individual in mmHg (unit)

5. Serum Cholestrol: displays the serum cholesterol in mg/dl (unit)

6. Fasting Blood Sugar: compares the fasting blood sugar value of an individual with 120mg/dl.
If fasting blood sugar > 120mg/dl then : 1 (true)
else : 0 (false)

7. Resting ECG : displays resting electrocardiographic results
0 = normal
1 = having ST-T wave abnormality
2 = left ventricular hyperthrophy

8. Max heart rate achieved : displays the max heart rate achieved by an individual.

9. Exercise induced angina :
1 = yes
0 = no

10. ST depression induced by exercise relative to rest: displays the value which is an integer or float.

11. Peak exercise ST segment :
1 = upsloping
2 = flat
3 = downsloping

12. Number of major vessels (0–3) colored by flourosopy : displays the value as integer or float.

13. Thal : displays the thalassemia :
3 = normal
6 = fixed defect
7 = reversible defect

14. Diagnosis of heart disease : Displays whether the individual is suffering from heart disease or not :
0 = absence
1 = present

```python
plt.figure(figsize=(15,12))

sns.countplot(x='age',data=data)

plt.title("Age Frequency",fontsize=20);
plt.xlabel("Age",fontsize=14)
plt.ylabel("Count",fontsize=14);
```

```python
plt.figure(figsize=(16,7))

sns.histplot(data.age[data.target == 1],kde=False,bins=50)

plt.title('Age of Heart Diseased Patients');
```

```python
fig, ax = plt.subplots(figsize=(20,20))

ax = fig.gca()

data.hist(ax=ax,bins=30)

plt.show();
```

```python
pd.crosstab(data.sex,data.target).plot(kind="bar", figsize=(10,5))

plt.title("Heart Disease Situation of Male and Female",fontsize=20)
plt.xlabel("Sex",fontsize=14)
plt.xticks(ticks=[0,1],labels=["Female","Male"])
plt.ylabel("Count",fontsize=14)
plt.legend(["No Heart Disease","Heart Disease"],fontsize=14);
```

```python
pd.crosstab(data.cp,data.target).plot(kind="bar",figsize=(10,8))

plt.title("Chest Pain Type vs Target",fontsize=20)
plt.xlabel("CP",fontsize=14)
plt.xticks(ticks=[0,1,2,3],labels=["typical angina","atypical angina"," non — anginal"," asymptotic"],fontsize=12)
plt.ylabel("Count",fontsize=14)
plt.legend(["No Heart Disease","Heart Disease"],fontsize=14);
```

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.scatter(data.age[data.target == 0],data.thalach[data.target == 0])
ax.scatter(data.age[data.target == 1],data.thalach[data.target == 1])

plt.title("Heart Disease in function of Age and Max Heart Rate",fontsize=20)
plt.xlabel("Age",fontsize=14)
plt.ylabel("Max Heart Rate",fontsize=14)
plt.legend([ "No Disease","Disease"],fontsize=14);
```

```python
plt.figure(figsize=(15,12))

sns.countplot(data= data, x='sex',hue='cp')

plt.title("")
plt.xlabel("Sex",fontsize=14)
plt.xticks(ticks=[0,1],labels=["Female","Male"],fontsize=10)
plt.ylabel("Count",fontsize=14)
plt.legend(["typical angina","atypical angina"," non — anginal"," asymptotic"],fontsize=12);
```

```python
df = data[data['target'] == 0 ][['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']]


ax = sns.jointplot(data=df,
              x='chol',
              y='trestbps',
              height=10
             )

ax.fig.suptitle("Heart Disease Patient Cholestrol and Blood Pressure Situation",fontsize=14)

ax.fig.tight_layout()
ax.set_axis_labels("Cholestrol", "Resting Blood Pressure", fontsize=14);
```

```python
plt.figure(figsize=(15,10))

sns.violinplot(data=data,x='target',y='thal')

plt.title("Thalassemia Situation of Patient",fontsize=20)
plt.xlabel("Diagnosis of Heart Disease",fontsize=14)
plt.xticks(ticks=[0,1],labels=["No Disease","Disease"],fontsize=12)
plt.ylabel("Thalassemia",fontsize=14);
```
