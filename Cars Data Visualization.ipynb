```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Dataset/Cars Data.csv")

data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.Make.value_counts()
```

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')
```

```python
df = data[data["Origin"] == "Europe"]
```

```python
plt.figure(figsize=(20,12))

sns.countplot(data=df,x="Make",hue="Origin")

plt.title("Made in Europe Cars",fontsize=20)
plt.xlabel("Make",fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel("Count",fontsize=14);
```
