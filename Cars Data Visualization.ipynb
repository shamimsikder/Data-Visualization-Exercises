```python
data.MSRP = data.MSRP.replace('[\$,\.]', '',regex=True).astype(int)

data.Invoice = data.Invoice.replace('[\$,\.]', '',regex=True).astype(int)

data
```

```python
data.Invoice.value_counts()
```

```python
data.MSRP.value_counts()
```

```python
eu = data[data.Origin == 'Europe']

asia = data[data.Origin == 'Asia']

usa = data[data.Origin == 'USA']
```

```python
eu_cars = eu.groupby("Type",as_index=False)

eu_cars = eu_cars.agg(np.mean)

asia_cars = asia.groupby("Type",as_index=False)

asia_cars = asia_cars.agg(np.mean)

usa_cars = usa.groupby("Type",as_index=False)

usa_cars = usa_cars.agg(np.mean)
```

```python
eu_cars.Invoice = eu_cars.Invoice.astype(int)

eu_cars.MSRP = eu_cars.MSRP.astype(int)

eu_cars
```

```python
asia_cars.Invoice = asia_cars.Invoice.astype(int)

asia_cars.MSRP = asia_cars.MSRP.astype(int)

asia_cars
```

```python
usa_cars.Invoice = usa_cars.Invoice.astype(int)

usa_cars.MSRP = usa_cars.MSRP.astype(int)

usa_cars
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,figsize=(15,20))

ax1=sns.barplot(hue=eu_cars.Type,x=eu_cars.Type,y=eu_cars.MSRP,ax=ax1)
ax2=sns.barplot(hue=asia_cars.Type,x=asia_cars.Type,y=asia_cars.MSRP,ax=ax2)

fig.suptitle("European and Asian Different Types of Cars Average MSRP",fontsize=20)

ax1.set_title("European Cars Average MSRP",fontsize=15,fontweight='roman',fontname="Sans")
ax2.set_title("Asian Cars Average MSRP",fontsize=15,fontweight='roman',fontname="Sans")

ax1.set_xlabel("Type",fontsize=12,fontweight='roman',fontname="Sans")
ax2.set_xlabel("Type",fontsize=12,fontweight='roman',fontname="Sans")

ax1.set_ylabel("MSRP",fontsize=12,fontweight='roman',fontname="Sans")
ax2.set_ylabel("MSRP",fontsize=12,fontweight='roman',fontname="Sans");
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,figsize=(15,20))

ax1 = sns.barplot(x=usa_cars.Type, y=usa_cars.MSRP, hue=usa_cars.Type, ax=ax1)
ax2 = sns.barplot(x=eu_cars.Type, y=eu_cars.MSRP, hue=eu_cars.Type, ax=ax2)

fig.suptitle("American and European Different Types of Cars Average MSRP",fontsize=20)

ax1.set_title("American Cars Average MSRP",fontsize=15,fontweight='roman',fontname="Sans")
ax2.set_title("European Cars Average MSRP",fontsize=15,fontweight='roman',fontname="Sans")

ax1.set_xlabel("Type",fontsize=12,fontweight='roman',fontname="Sans")
ax2.set_xlabel("Type",fontsize=12,fontweight='roman',fontname="Sans")

ax1.set_ylabel("MSRP",fontsize=12,fontweight='roman',fontname="Sans")
ax2.set_ylabel("MSRP",fontsize=12,fontweight='roman',fontname="Sans");
```
