```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Dataset/Cars Data.csv")

data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.Make.value_counts()
```

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')
```

```python
data.Origin.value_counts()
```

```python
df = data[data["Origin"] == "Europe"]
df2 = data[data.Origin == "Asia"]
df3 = data[data.Origin == "USA"]
```

```python
plt.figure(figsize=(15,10))

sns.countplot(data=df,x="Make",hue="Origin")

plt.title("Made in Europe Cars",fontsize=18)
plt.xlabel("Make",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Count",fontsize=14);
```

```python
plt.figure(figsize=(15,10))

sns.countplot(data=df2,x="Make",hue="Origin")

plt.title("Made in Asia Cars",fontsize=18)
plt.xlabel("Makes",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Count",fontsize=14);
```

```python
plt.figure(figsize=(15,10))

sns.countplot(data=df3,x="Make",hue="Origin")

plt.title("Made in USA Cars",fontsize=18)
plt.xlabel("Makes",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Count",fontsize=14);
```

```python
plt.figure(figsize=(15,10))

sns.countplot(data=df,x="Make",hue="Type")

plt.title("Type of Cars Europe Made",fontsize=18)
plt.xlabel("Makes",fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Count",fontsize=14);
```

```python
plt.figure(figsize=(30,20))

sns.countplot(data=df2,x="Make",hue="Type")

plt.title("Type of Cars Asia Made",fontsize=28)
plt.xlabel("Makes",fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20, loc=2)
plt.ylabel("Count",fontsize=20);
```

```python
plt.figure(figsize=(30,20))

sns.countplot(data=df3,x="Make",hue="Type")

plt.title("Type of Cars USA Made",fontsize=28)
plt.xlabel("Makes",fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.ylabel("Count",fontsize=20);
```

```python
plt.figure(figsize=(15,10))

sns.scatterplot(data=df, x="Make", y="MPG_City",hue="Type", marker="o",s=50)

plt.title("European Cars MPG in City(Type Wise)",fontsize=20)
plt.xlabel("Makes",fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12,loc=2)
plt.ylabel("Miles Per Gallon(MPG)",fontsize=15);
```
